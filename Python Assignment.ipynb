# Data handling

```python
df.Degree.unique()#no need to handle
```

```python
#handling specalization column
df.Specialization.unique()
```

```python
df.Specialization.value_counts()
```

less numbers are not good for creating a model. so i put all the values less than 10 into other column

```python
# assigning values into other category
spec=df.Specialization.value_counts()
spec
```

```python
#get specialization less than 10
specless10=spec[spec<=10]
specless10
```

```python
def filter(value):
    if value in specless10:
        return 'other'
    else:
        return value
    
df.Specialization=df.Specialization.apply(filter)
df.Specialization.unique()
```

in domain there is -1 value.so i decided to remove that and replace it with a mean value.others are good.

```python
df=df.replace(-1,np.nan)
```

```python
#assign null values with thier mean
cols_with_nan = [column for column in df.columns if df.isna().sum()[column] > 0]
for column in cols_with_nan:
    df[column] = df[column].fillna(df[column].mean())
```
